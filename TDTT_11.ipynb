```python
def tongbinhphuong(n):
    sum=0
    while n>0:
        chuso=n%10
        sum+=chuso**2
        n//=10
    return sum
def happynumber(n):
    save=set()
    while n!=1:
        if n in save :
            return False
        save.add(n)
        n=tongbinhphuong(n)
    return True
try:
    n=int(input())
    result=happynumber(n)
    if result:
        print("YES")
    else:
        print("NO")
except ValueError:
    print("Loi nhap khong phai so nguyen")
```

```python
def solve(s):
    ngoactron=0
    lap=False
    i=0
    while i<len(s):
        char=s[i]
        if char=='\\':
            if i+1>=len(s):return False
            lap=True
            i+=2
            continue
        if char=='[':
            dong=s.find(']',i)
            if dong==-1:return False
            lap=True   
            i=dong+1
            continue
        if char in '*+?':
            if not lap:return False
            lap=False
            i+=1
            continue
        if char=='(':
            ngoactron+=1
            lap=False
        elif char==')':
            ngoactron-=1
            if ngoactron<0:return False
            lap=True
        else:
            lap=True
        i+=1
    return ngoactron==0
try:
    T=int(input())
    for _ in range(T):
        s=input()
        if solve(s):
            print("True")
        else:
            print("False")
except ValueError:
    pass
            
```
