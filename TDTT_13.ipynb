```python
#W11A1
def reverse(path: str) -> int:
    with open(path,'r',encoding='utf-8') as file:
        n=file.read().strip()
    reverse_n=n[::-1]
    return int(reverse_n)
```

```python
#W11A4
def areAnagrams(path1: str, path2: str):
    def check(text):
        text=text.lower()
        save=[i for i in text if i.isalpha()]
        return sorted(save)
    with open(path1,'r',encoding='utf-8')as file1:
        f1=file1.read()
    with open(path2,'r',encoding='utf-8') as file2:
        f2=file2.read()
    return check(f1)==check(f2)
    
```

```python
#W11A11
def find(path: str):
    try:
        with open(path, 'r', encoding='utf-8') as file:
            data = file.read().split()
            if not data:
                return
            numbers = [int(x) for x in data]
            max_val = max(numbers)
            min_val = min(numbers)
            print(f"{max_val} {min_val}")
    except Exception:
        print("Mission failed")
```

```python
#W11A12
def maximumProduct (path: str) -> int: 
    with open(path,'r',encoding='utf-8') as file:
        n=int(file.readline().strip())
        nums=[int(x) for x in file.readline().split()]
    nums.sort()
    case1 = nums[-1] * nums[-2] * nums[-3]
    case2 = nums[0] * nums[1] * nums[-1]
    return max(case1, case2)
```

```python
#W11A13
def moveZeroes (fileName: str) -> list :
    with open(fileName, 'r', encoding='utf-8') as f:
        n_line = f.readline()
        if not n_line: return []
        n = int(n_line.strip())
        nums = [int(x) for x in f.readline().split()]
    non_zeros = [x for x in nums if x != 0]
    zero_count = len(nums) - len(non_zeros)
    result = non_zeros + ([0] * zero_count)
    return result
```

```python
#W11A14
import csv
def averageTime(path: str) -> float:
    with open(path, 'r', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        total_time = 0.0
        count = 0
        for row in reader:
            try:
                time_val = float(row['Time'])
                total_time += time_val
                count += 1
            except ValueError:
               continue
        if count == 0:
            return 0.0
    return round(total_time / count, 2)
```

```python
#W11A115
import csv
def findMovies(path: str) -> list[str]:
    result_movies = []
    with open(path, 'r', encoding='utf-8') as f:
        reader = csv.DictReader(f, skipinitialspace=True)
        for row in reader:
            try:
                movie_name = row['movie']
                description = row['description']
                rating = float(row['rating'])
                if rating > 8.0 and description.lower() != 'boring':
                    result_movies.append(movie_name)
            except ValueError:
                continue
    if not result_movies:
        return ["All are boring!"]
    return result_movies
```

```python
#W11A16
import math
import sys
def calculateperimeters():
    file_path = input().strip()
    try:
        with open(file_path, 'r') as f:
            lines = [line.strip() for line in f.readlines() if line.strip()]
        i = 0
        while i < len(lines):
            line = lines[i]
            if line.startswith("TYPE:"):
                shape_type = line.split(":")[1].strip()
                if i + 1 < len(lines):
                    params = lines[i+1].split()
                    perimeter = 0.0
                    if shape_type == "SQUARE":
                        side = float(params[1])
                        perimeter = 4 * side
                    elif shape_type == "RECTANGLE":
                        width = float(params[1])
                        height = float(params[3])
                        perimeter = 2 * (width + height)
                    elif shape_type == "CIRCLE":
                        radius = float(params[1])
                        perimeter = 2 * math.pi * radius
                    print(f"{perimeter:.2f}")
                    i += 2
                else:
                    break
            else:
                i += 1
    except FileNotFoundError:
        print("Không tìm thấy tệp.")
    except Exception as e:
        print(f"Đã xảy ra lỗi: {e}")
```

```python
#W11A17
def productExceptSelf(path: str) -> list[int]:
    with open(path, 'r', encoding='utf-8') as f:
        nums = [int(x) for x in f.read().split()]
    n = len(nums)
    if n <= 1:
        return []
    answer = [1] * n
    leftproduct = 1
    for i in range(n):
        answer[i] = leftproduct
        leftproduct *= nums[i]
        rightproduct = 1
    for i in range(n - 1, -1, -1):
        answer[i] *= rightproduct
        rightproduct *= nums[i]
    print(answer)
    return answer
```

```python
#W11A18
import math
def estimatedTime(path: str):
    try:
        with open(path, 'r') as f:
            lines = f.readlines()
        if len(lines) < 3:
            return -1
        parts_src = lines[0].strip().split()
        x1 = float(parts_src[2])  
        y1 = float(parts_src[4])  
        parts_des = lines[1].strip().split()
        x2 = float(parts_des[2])
        y2 = float(parts_des[4])
        parts_vel = lines[2].strip().split()
        velocity = float(parts_vel[1])
        distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
        if velocity <= 0:
            return 0
        time = distance / velocity
        return round(time, 2)
    except FileNotFoundError:
        print("Không tìm thấy tệp tin.")
        return -1
    except Exception as e:
        print(f"Lỗi định dạng dữ liệu: {e}")
        return -1
```

```python
#W11A19
import struct
import sys
def read_binary_doubles():
    file_path = input().strip()
    try:
        with open(file_path, 'rb') as f:
            while True:
                bytechunk = f.read(8)
                if not bytechunk:
                    break
                if len(bytechunk) < 8:
                    break
                value = struct.unpack('d', bytechunk)[0]
                print(f"{value:.5f}")
    except FileNotFoundError:
        print("Không tìm thấy tệp tin.")
    except Exception as e:
        print(f"Đã xảy ra lỗi: {e}")
```

```python
#W11A20
import math
def imageSmoother(path: str):
    try:
        with open(path, 'r') as f:
            content = f.read().split()        
        if not content:
            return
        iterator = iter(content)      
        try:
            M = int(next(iterator))
        except StopIteration:
            return
        img = []
        for _ in range(M):
            row = []
            for _ in range(M):
                row.append(int(next(iterator)))
            img.append(row)
        result = [[0] * M for _ in range(M)]
        directions = [
            (-1, -1), (-1, 0), (-1, 1),
            (0, -1),  (0, 0),  (0, 1),
            (1, -1),  (1, 0),  (1, 1)
        ]
        for r in range(M):
            for c in range(M):
                total_sum = 0
                count = 0
                for dr, dc in directions:
                    nr, nc = r + dr, c + dc
                    if 0 <= nr < M and 0 <= nc < M:
                        total_sum += img[nr][nc]
                        count += 1
                result[r][c] = total_sum // count
        for row in result:
            print(" ".join(map(str, row)))

    except FileNotFoundError:
        print("Không tìm thấy tệp tin.")
    except Exception as e:
        print(f"Lỗi: {e}")
```
