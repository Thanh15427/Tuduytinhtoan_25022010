```python
#W5A1
def Max_Of_Two(a: int, b):
    if a > b:
        return a
    else:
        return b
a,b=map(int,input().split())
print(Max_Of_Two(a,b))
```

```python
#W5A2
def swap(a,b):
    return b,a
a,b=map(int,input().split())
print(*swap(a,b))
```

```python
#W5A3
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
num = int(input())
print(is_prime(num))
```

```python
#W5A4
def is_perfect_number(n):
    if n < 1:
        return False
    sum_of_divisors = 0
    for i in range(1, n):
        if n % i == 0:
            sum_of_divisors += i
    return sum_of_divisors == n
num = int(input())
print(is_perfect_number(num))
```

```python
#W5A5
def find_first_occurrence(arr, k):
    for i in range(len(arr)):
        if arr[i] == k:
            return i+1
    return -1
n = list(map(int, input().split()))
k = int(input())
print(find_first_occurrence(n, k))
```

```python
#W5A6
def factorial(n):
    if n < 0:
        return None
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result
n = int(input())
print(factorial(n))
```

```python
#W5A7
def calculator(num1, num2, operat):
    if operat == '+':
        return round(num1 + num2, 2)
    elif operat == '-':
        return round(num1 - num2, 2)
    elif operat == '*':
        return round(num1 * num2, 2)
    elif operat == '/':
        if num2 != 0:
            return round(num1 / num2, 2)
        else:
            return "Error"
    else:
        return "Error"
num1,operat,num2=input().split()
num1,num2=float(num1),float(num2)
print(calculator(num1, num2, operat))
```

```python
#W5A8
def hamming(a,b):
    xor=a^b
    count=0 
    while xor:
        count+=xor&1
        xor>>=1
    return count       
a,b=map(int,input().split())
print(hamming(a,b))
```

```python
#W5A9
def sum_of_digits(n):
    if n < 0:
        return 0
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    return total
n = int(input())
print(sum_of_digits(n))
```

```python
#W5A10
def checks2(a,b):
    if len(a)!=len(b):return False
    char_a_to_b={}
    char_b_to_a={}
    for i in range(len(a)):
        char_a=a[i]
        char_b=b[i]
        if (char_a in char_a_to_b and char_a_to_b[char_a]!=char_b) or (char_b in char_b_to_a and char_b_to_a[char_b]!=char_a):
            return False
        char_a_to_b[char_a]=char_b
        char_b_to_a[char_b]=char_a
    return True
a,b=input().split() 
print(checks2(a,b))
```
