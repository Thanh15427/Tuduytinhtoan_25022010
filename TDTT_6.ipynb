```python
#W6A9
dong1=input().split()
dong2=input().split()
d1={dong1[i]:int(dong1[i+1]) for i in range(0,len(dong1),2)}
d2={dong2[i]:int(dong2[i+1]) for i in range(0,len(dong2),2)}
d3={}
for key in d1:
    if key in d2:
        d3[key]=d1[key]+d2[key]
    elif key not in d2:
        d3[key]=d1[key]
for key in d2:
    if key not in d1:
        d3[key]=d2[key]
print(d3)
```

```python
#W6A10
day=list(map(int,input().split()))
k=int(input())
d=[]
result=set()
for i in day:
    target=k-i
    if target in d:
        result.add((target,i))
    d.append(i)
d = sorted(list(result))
print(d)
```

```python
#W6A16
n=int(input("nhap so phan tu tren hang:"))
m=int(input("nhap so phan tu tren cot:"))
matrix=[]
for i in range(n):
    row=list(map(int,input().split()))
    if len(row)!=m:
        print("So phan tu tren cot khong hop le!")
        exit()
    else:
        matrix.append(row)
for row in matrix:
    for so in row:
        print(f"{so:>4}",end="")
    print()
```

```python
#W6A17
n=int(input("Nhap so hang,cot ma tran vuong:"))
matrix=[]
for i in range(n):
    row=list(map(int,input().split()))
    if len(row)!=n:
        print("So phan tu tren cot khong hop le!")
        exit()
    else:
        matrix.append(row)
duongcheochinh=[]
duongcheophu=[]
for i in range(n):
    duongcheochinh.append(matrix[i][i])
    duongcheophu.append(matrix[i][n-1-i])
print("Duong cheo chinh:",duongcheochinh)
print("Duong cheo phu:",duongcheophu)   
```

```python
#W6A18
n,m,k=map(int,input().split())
matrix=[]
for i in range(n):
    row=list(map(int,input(f"Nhap hang {i+1}: ").split()))
    if len(row)!=m:
        print("Khong thoa man")
        exit()
    else:
        matrix.append(row)
tongcot=0
if k<=n and k<=m:
    tonghang=sum(matrix[k-1])
    for j in range(n):
        tongcot+=matrix[j][k-1]
    print(tonghang)
    print(tongcot)
else:
    print("Khong thoa man")
```
