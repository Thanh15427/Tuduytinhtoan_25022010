```python
#W7A1
def binary_search(arr, left,right, target):
    left=0
    right=len(arr)-1
    while left<=right:
        mid=(left+right)//2
        if arr[mid]==target:
            return mid
        elif arr[mid]<target:
            left=mid+1
        else:
            right=mid-1
    return -1
arr=list(map(int,input().split()))
target=int(input())
left=0
right=len(arr)-1
print(binary_search(arr,left,right,target))
```

```python
#W7A3
#Bubble sort
def bubble_sort(arr):
    n=len(arr)
    for i in range(n):
        swapped = False
        for j in range(0,n-i-1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
                swapped=True
        if not swapped :
            break
    return arr
arr=list(map(int,input().split()))
print(bubble_sort(arr))
#Selection sort
def selection_sort(arr):
    n=len(arr)
    for i in range(n):
        min_arr=i
        for j in range(i+1,n):
            if arr[j]<arr[min_arr]:
                arr[j],arr[min_arr]=arr[min_arr],arr[j]
    return arr
arr=list(map(int,input().split()))
print(selection_sort(arr))
```

```python
#W7A5
def count_pairs_with_sum(arr,x):
    seen={}
    count=0
    for num in arr:
        needed=x-num
        if needed in seen:
            count+=seen[needed]
        if num in seen:
            seen[num]+=1
        else:
            seen[num]=1
    return count
arr=list(map(int,input().split()))
x=int(input())
print(count_pairs_with_sum(arr,x))
```

```python
#W7A7
def lenght_of_list(nums):
    sub=[]
    for i in nums:
        if not sub or i >sub[-1]:
            sub.append(i)
        else:
            idx=find_index(sub,i)
            sub[idx]=i
    return len(sub)
def find_index(arr,x):
    left=0
    right=len(arr)-1
    result=len(arr)
    while left<=right:
        mid=(left+right)//2
        if arr[mid]>=x:
            result=mid
            right=mid-1
        else:
            left=mid+1
    return result
a=list(map(int,input().split()))
print(lenght_of_list(a))
```

```python
#W7A9
def binary_search(arr,x):
    left=0
    right=len(arr)
    while left<right:
        mid=(left+right)//2
        if arr[mid]<x:
            left=mid+1
        else:
            right=mid
    return left
def count_smaller_manual(nums):
    sorted=[]
    result=[]
    for i in reversed(nums):
        idx = binary_search(sorted,i)
        result.append(idx)
        sorted.insert(idx,i)
    return result[::-1]
nums=list(map(int,input().split()))
print(count_smaller_manual(nums))
```

```python
#W7A14
def search_seven_memory(n, arr):
    found = False
    result = []
    for i in range(n - 1, -1, -1):
        if arr[i] == 7:
            result.append(str(i))
            found = True
    if not found:
        return "Not found"
    return " ".join(result)
n = int(input())
arr = list(map(int, input().split()))
print(search_seven_memory(n, arr))
```
