```python
#W8A1
class cylinder():
    def __init__ (self,radius,height):
        self.radius=radius
        self.height=height
    def circumference(self):
        return f"{2*3.14*(self.radius)**2+self.height*2*3.14*self.radius:.2f}"
    def volume(self):
        return f"{3.14*((self.radius)**2)*self.height:.2f}"
a=list(map(float,input().split()))
b=cylinder(a[0],a[1])
print(b.circumference())
print(b.volume())
```

```python
#W8A2
class Date():
    def __init__(self,day,month,year):
        self.day=day
        self.month=month
        self.year=year
    def is_leap_year(self):
        return self.year%4==0 and self.year%100 !=0 or self.year%400==0
    def days_in_month(self):
        if self.month in [1,3,5,7,8,10,12]:
            return 31
        elif self.month in [4,6,9,11]:
            return 30
        elif self.month ==2:
            return 29 if self.is_leap_year() else 28
        else:
            return 0
    def is_valid(self):
        if self.year < 0:
            return False
        if self.month < 1 or self.month > 12:
            return False    
        max_day = self.days_in_month()
        if self.day < 1 or self.day > max_day:
            return False
        return True
    def next_day(self):
        self.day+=1
        if self.day >self.days_in_month():
            self.day=1
            self.month+=1
            if self.month >12:
                self.month=1
                self.year+=1
    def __str__(self):
        return f"{self.day}/{self.month}/{self.year}"
d, m, y = map(int, input().split('/'))
date_obj = Date(d, m, y)
if date_obj.is_valid():
    date_obj.next_day()
    print(date_obj)
else:
    print("INVALID")
```

```python
#W8A3
class Point():
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def check(self):
        if self.x==0 and self.y==0:
            return "M is origin"
        elif self.x==0 and self.y!=0:
            return "M located on Y-axis"
        elif self.x!=0 and self.y==0:
            return "M located on X-axis"
        else :
            dist=((self.x)**2+(self.y)**2)**0.5
            return f"d(M,O)={dist:.2f}"
x,y=map(int,input().split())
point=Point(x,y)
print(point.check())
```

```python
#W8A4
class Calculator():
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def add(self):
        return self.a+self.b
    def subtract(self):
        return self.a-self.b
    def product(self):
        return self.a*self.b
    def divide(self):
        if self.b==0:
            return "Error"
        else:
            return self.a/self.b
    def power(self):
        return self.a**self.b
    def mod(self):
        if self.b==0:
            return "Error"
        else:
            return self.a%self.b
    def set_numbers(self,a,b):
        self.a=a
        self.b=b
        print(f"Đã cập nhật số mới: {self.a} và {self.b}")
print("-----Calculator-----")
n1=float(input("Nhập số thứ nhất:"))
n2=float(input("Nhập số thứ hai :"))
calc=Calculator(n1,n2)
while True:
    print("\n--- MENU ---")
    print("1. Cộng (add)")
    print("2. Trừ (subtract)")
    print("3. Nhân (product)")
    print("4. Chia (divide)")
    print("5. Lũy thừa (power)")
    print("6. Chia lấy dư (mod)")
    print("7. Nhập 2 số mới (set_numbers)")
    choice = input("Chọn phép toán (1-7): ")
    if choice == '1':
        print(f"{calc.add()}")
    elif choice == '2':
        print(f"{calc.subtract()}")
    elif choice == '3':
        print(f"{calc.product()}")
    elif choice == '4':
        print(f"{calc.divide()}")
    elif choice == '5':
        print(f"{calc.power()}")
    elif choice == '6':
        print(f"{calc.mod()}")
    elif choice == '7':
        new_n1 = float(input("Nhập số thứ nhất mới: "))
        new_n2 = float(input("Nhập số thứ hai mới: "))
        calc.set_numbers(new_n1, new_n2)
    else:
            print("Lựa chọn không hợp lệ!")
    ask=input("Bạn có muốn thoát?")
    if ask in ["yes", "YES","Yes"]:
        print("Đã thoát chương trình.")
        break
```

```python
#W8A5
class ShoppingCart():
    def __init__(self):
        self.cart=[]
    def add(self,name:str,price:float):
        item={
            "name":name,
            "price":price
        }
        self.cart.append(item)
        print(f"Đã thêm {name} với giá {price}")
    def remove(self,name:str):
        found=False
        for item in self.cart[:]:
            if item["name"]==name:
                self.cart.remove(item)
                print(f"Đã xóa {name} khỏi giỏ hàng")
                found=True
                break
        if not found:
            print(f"{name} không có trong giỏ hàng")
    def empty(self):
        return len(self.cart)==0
    def total(self):
        total=sum(item["price"] for item in self.cart)
        return total
    def display(self):
        if self.empty():
            print("Giỏ hàng đang rỗng")
        else:
            print("--- ShoppingCart ---")
            for i,item in enumerate(self.cart,1):
                print(f"{i}.{item['name']}-Giá:{item['price']}")
            print("--------------------")
    def clear(self):
        self.cart=[]
        print("Đã xóa toàn bộ giỏ hàng")
my_cart=ShoppingCart()
while True:
    print("\n================ MENU ================")
    print("2. Thêm sản phẩm vào giỏ")
    print("3. Xóa sản phẩm theo tên")
    print("4. Kiểm tra giỏ hàng rỗng")
    print("5. Tính tổng tiền")
    print("6. Hiển thị giỏ hàng")
    print("7. Xóa toàn bộ giỏ hàng")
    print("======================================")
    choice = input("Nhập lựa chọn của bạn (2-7): ")
    if choice=='2':
        name=input("Nhập tên sản phẩm:")
        price=float(input("Nhập giá sản phẩm:"))
        my_cart.add(name,price)
    elif choice=="3":
        name=input('Nhập tên sản phẩm cần xóa:')
        my_cart.remove(name)
    elif choice=="4":
        if my_cart.empty():
            print("Giỏ hàng đang rỗng")
        else:
            print("Sản phẩm đã trong giỏ hàng")
    elif choice =="5":
        total=my_cart.total()
        print(f"Tổng số tiền : {total}")
    elif choice =="6":
        my_cart.display()
    elif choice =="7":
        my_cart.clear()
    exit=input("Có muốn thoát khỏi chương trình không?(yes/no)")
    if exit.strip().lower =="yes":
        print("Cảm ơn đã sử dụng giỏ hàng")
        break
```

```python
#W8A6
class Rectangle():
    def __init__(self,w,h):
        self.w=w
        self.h=h
    def area(self):
        return f"{self.w*self.h:.2f}"
    def perimeter(self):
        return f"{(self.w+self.h)*2:.2f}"
    def scale(self,k):
        self.w*=k
        self.h*=k
w,h,k=map(float,input().split())
result=Rectangle(w,h)
result.scale(k)
print(f"{result.area()} {result.perimeter()}")
```

```python
#W8A7
class Fraction():
    def __init__(self,num,den):
        if den==0:
            raise ValueError("Mẫu số không thể bằng 0")
        if den<0 :
            num=-num
            den=-den
        common=self.ucln(num,den)
        self.num=num//common
        self.den=den//common
    def ucln(self,a,b):
        a,b= abs(a) ,abs(b)
        while b!=0:
            a,b=b,a%b
        return a
    def add(self, other):
        new_num = self.num * other.den + self.den * other.num
        new_den = self.den * other.den
        return Fraction(new_num, new_den)
    def sub(self, other):
        new_num = self.num * other.den - self.den * other.num
        new_den = self.den * other.den
        return Fraction(new_num, new_den)
    def mul(self,other):
        new_num=self.num*other.num
        new_den=self.den*other.den
        return Fraction(new_num,new_den)
    def div(self,other):
        new_num=self.num*other.den
        new_den=self.den*other.num
        return Fraction(new_num,new_den)
    def __str__(self):
        return f"{self.num}/{self.den}"
input_data=input().split()
if len(input_data) == 5:
    a=int(input_data[0])
    b=int(input_data[1])
    op=input_data[2]
    c=int(input_data[3])
    d=int(input_data[4])
    f1=Fraction(a,b)
    f2=Fraction(c,d)
    result=None
    if op=="+":
        result=f1.add(f2)
    elif op=="-":
        result=f1.sub(f2)
    elif op=="*":
        result=f1.mul(f2)
    elif op=="/":
        result=f1.div(f2)
    if result:
        print(result)
        
```

```python
#W8A8
class BankAccount():
    def __init__(self,owner,blance):
        self.owner=owner
        self.blance=float(blance)   
    def deposit(self,amount):
        self.blance+=amount
    def withdraw(self,amount):
        if amount<=self.blance:
            self.blance-=amount
    def __str__(self):
        return f"{self.owner}:{self.blance}"
Data=input().split()
a=Data[0]
b=float(Data[1])
my_account=BankAccount(a,b)
n=int(input())
for _ in range(n):
    command=input().split()
    read_cmd=command[0]
    amount=float(command[1])
    if read_cmd == "DEPOSIT":
        my_account.deposit(amount)
    elif read_cmd =="WITHDRAW":
        my_account.withdraw(amount)
print(f"{my_account.blance:.2f}")
```
