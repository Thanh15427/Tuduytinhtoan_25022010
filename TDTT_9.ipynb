```python
#W9A1
def merge_sort_rec(a,b):
    if not a:
        return b
    if not b:
        return a
    if a[0]<=b[0]:
        return [a[0]]+merge_sort_rec(a[1:],b)
    else:
        return [b[0]]+merge_sort_rec(b[1:],a)
a=list(map(int,input().split()))
b=list(map(int,input().split()))
print(merge_sort_rec(a,b))
```

```python
#W9A2
def count1(matrix,mid):
    n,m=len(matrix),len(matrix[0])
    row,col=0,m-1
    count=0
    while row<n and col>=0:
        if matrix[row][col]>=mid:
            count+= n-row
            col-=1
        else:
            row+=1
def binary_search(low,high,matrix,k):
    if low==high:
        return low
    mid=(low+high)//2
    count=count1(matrix,mid)
    if count>=k:
        return binary_search(mid,high,matrix,k)
    else:
        return binary_search(low,mid-1,matrix,k)
a=list(map(int,input().split()))
b=list(map(int,input().split()))
c=list(map(int,input().split()))
matrix=[]
matrix.append(a)
matrix.append(b)
matrix.append(c)
k=int(input())
print(binary_search(matrix[0][0],matrix[len(matrix)-1][len(matrix)-1],matrix,k))
```

```python
#W9A3
def solve(s,index):
    if index>=len(s)-1:
        return s
    min=s[index]
    target=-1
    for i in range(len(s)-1,index,-1):
        if s[i]<min:
            min=s[i]
            target=i
    if target!=-1:
        s[index],s[target]=s[target],s[index]
        return s
    return solve(s,index+1)
s=input()
chars=list(s)
a=solve(chars,0)
result="".join(a)
print(int(result))
```

```python
#W9A4
def solve(a,b):
    if len(b)==0:
        return a.islower() or a==" "
    if len(a)==0:
        return False
    chara=a[0]
    charb=b[0]
    if chara.isupper():
        if chara == charb:
            return solve(a[1:],b[1:])
        else:
            return False
    else:
        delete=solve(a[1:],b)
        save=False
        if chara.upper()==charb:
            save=solve(a[1:],b[1:])
        return delete or save
n=int(input())
for i in range(n):
    a=input()
    b=input()
    if solve(a,b):print("YES")
    else:print("NO")
```

```python
#W9A5
def is_sublist_rec(a,b):
    if len(a)>len(b):
        return "NO"
    if b[:len(a)]==a:
        return "YES"
    return is_sublist_rec(a,b[1:])
list1=list(map(int,input().split()))
list2=list(map(int,input().split()))
print(is_sublist_rec(list1,list2))
```
