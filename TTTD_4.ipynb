```python
#[LargestPrimeFactor]
n= int(input()) 
max_prime = -1
while n % 2 == 0:
    max_prime = 2
    n //= 2
for i in range(3, int(n**0.5) + 1, 2):
    while n % i == 0:
        max_prime = i
        n //= i
if n > 2:
    max_prime = n
print(max_prime)
```

```python
#[PerfectSquaresDistinctDigits]
n=int(input())
tim_so=False
i=1
while i*i<n:
    so_chinh_phuong=i*i
    chuoi_so=str(so_chinh_phuong)
    if len(set(chuoi_so))==len(chuoi_so):
        print(so_chinh_phuong,end=" ")
        tim_so=True
    i+=1
if not tim_so:
    print("No number")
```

```python
#[LongestCollatzUpToN]
n=int(input())
so_ket_qua = 0
max_do_dai = 0
for i in range(1, n + 1):
    x = i
    do_dai_hien_tai = 1
    while x != 1:
        if x % 2 == 0:
            x //= 2
        else:
            x = 3 * x + 1
        do_dai_hien_tai += 1
    if do_dai_hien_tai > max_do_dai:
        max_do_dai = do_dai_hien_tai
        so_ket_qua = i
print(so_ket_qua)    
print(max_do_dai)
```

```python
#W4A34
a = input("Nhập chuỗi a: ")
b = input("Nhập chuỗi b: ")
ket_qua = ""
i = 0
do_dai_a = 0
do_dai_b = 0
for _ in a:
    do_dai_a += 1
for _ in b:
    do_dai_b += 1
while i < do_dai_a:
    match = True
    if i + do_dai_b <= do_dai_a:
        for j in range(do_dai_b):
            if a[i + j] != b[j]:
                match = False
                break
        if match:
            i += do_dai_b
            continue
    ket_qua += a[i]
    i += 1
print(ket_qua)
```
